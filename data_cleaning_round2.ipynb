```python
def parse_range(value):
    # Define the regex pattern to match "<number> <unit>"
    pattern = re.compile(r'(\d+(?:\.\d+)?)\s*(L|Cr)')

    # Find all matches
    matches = pattern.findall(value)

    if len(matches) != 2:
        raise ValueError("Input does not contain a valid range.")

    # Parse the matches
    start_value, start_unit = matches[0]
    end_value, end_unit = matches[1]

    # Convert to float
    start_value = float(start_value)
    end_value = float(end_value)

    # Convert to integer based on unit
    if start_unit == 'L':
        start_value *= 100_000
    elif start_unit == 'Cr':
        start_value *= 10_000_000

    if end_unit == 'L':
        end_value *= 100_000
    elif end_unit == 'Cr':
        end_value *= 10_000_000

    return (start_value+end_value)/2
```

```python
def extract_and_average(text):
    # Find all patterns like 'number - number X'
    if '-' in str(text): # and 'Cr' in str(text) and 'L' not in str(text):
      matches = re.findall(r'(\d*\.?\d+)\s*-\s*(\d*\.?\d+)\s*([LCr]+)', text)
      results = []

      for match in matches:
          num1 = float(match[0])
          num2 = float(match[1])
          suffix = match[2]
          average = (num1 + num2) / 2
          results.append(f"{average:.2f} {suffix}")

      return results[0]
    else:
      return text
```

```python
def get_landmark_count(s):
    # landmark = ['Railway Station','Metro Station','Airport','Airports','Connectivity','Metro Stations','Railway Stations','Bus Depots','Connectivities','Bus Stops','Bus Stop','Bus Depot']
    # landmark = ['Hospital','Hospitals','Pharmacys','Pharmacy']
    # landmark = ['Religious Places','Religious Place']
    # landmark = ['Educations','Education','Library','Libraries']
    # landmark = ['ATM','ATMs','Bank','Banks','Bankss']
    # landmark = ['Shopping','Shoppings']
    landmark = ['Office Complex','Office Complexes']
    num = 0
    # Define the regex pattern to match "<number> <landmark>"
    pattern = re.compile(r'(\d+)\s([A-Za-z\s]+?)(?=,\s|$)')

    # Find all matches
    matches = pattern.findall(str(s))

    # Iterate over matches to find the specified landmark
    for count, name in matches:
        if name.strip() in landmark:
          num += int(count)
    return int(num)
```

```python
ind['office_complex'] = ind['landmark_details'].apply(get_landmark_count)
```
